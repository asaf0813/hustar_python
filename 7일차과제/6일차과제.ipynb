```python
import copy

class Set:
    def __init__(self, member=[]):
        self.member = member
        l = []
        for i in member:
            if i not in l:
                l.append(i)
        self.member = l
        
    def append(self, a):
        l = self.member        
        if a not in self.member:
            l.append(a)
        return l
    def delete(self, a):
        l = self.member
        if a in self.member:
            l.remove(a)
        return l    
    def union(self, s2):
        l = self.member
        temp = copy.deepcopy(l)
        for i in s2.member:
            if i in l:
                continue
            else :
                temp.append(i)
        return temp
    
    def intersection(self, s2):
        temp = []
        for i in self.member:
            if i in s2.member:
                temp.append(i)
        return temp
    
    def difference(self, s2):
        temp = []
        for i in self.member:
            if i not in s2.member:
                temp.append(i)
        return temp
    
    def __add__(self, member):
        a = self.union(member)
        return a
    def __sub__(self, member):
        a = self.difference(member)
        return a
    def __truediv__(self, member):
        a = self.intersection(member)
        return a
    def __repr__(self):
        return str(self.number)
    
a = Set([1, 2, 3])
b = Set([2, 3, 4])
c = a.append(4)
print(c)
# c = a.union(b)
# print(c)
# d = a.difference(b)
# print(d)
# e = a.intersection(b)    
# print(e)
# c = a + b
# print(c)
# d = a - b
# print(d)
# e = a / b
# print(e)
```
